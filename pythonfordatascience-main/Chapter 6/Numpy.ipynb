```python
df['surplus'] = df['income_mean'] - df['expenditure_mean']

df

# 这行代码计算每一行的 income_mean（平均收入）减去 expenditure_mean（平均支出）的差值，并将结果存储在一个新的列 surplus（盈余）中。
# 这可以帮助我们了解每个记录中收入和支出之间的差额，或者说是盈余。
```

```python
df[df["income_mean"]>10000]
```

```python
df.corr(numeric_only=True)["income_mean"]["surplus"]

# df.corr(numeric_only=True): 计算数据框中所有数值列之间的相关系数矩阵。numeric_only=True 参数表示仅考虑数值列。
# ["income_mean"]: 从相关系数矩阵中选择 income_mean 列。这个操作会返回一个包含 income_mean 与其他列相关系数的 Series。
# ["surplus"]: 从上一步返回的 Series 中选择 surplus 列。这会返回 income_mean 与 surplus 的相关系数值。
```

```python
# https://github.com/mwaskom/seaborn-data
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic

# 这行代码使用 seaborn 的 load_dataset 函数加载名为 tips 的示例数据集，并将其存储在 titanic 变量中。
# tips 数据集包含餐馆服务员的收小费情况，包括账单总额、性别、吸烟与否、就餐日期、就餐时间和聚餐人数等信息。
```

```python
titanic.info()
```

```python
titanic['age'][300:310]
```

```python
titanic['age'] = titanic.age.interpolate("nearest")
# 这行代码使用插值方法填充 titanic 数据集中 age（年龄）列的缺失值。具体来说，
# 它使用 nearest（最近邻）插值方法，这意味着用最近的非缺失值来填充缺失值。

titanic["age"][290:310]
# 这行代码打印 titanic 数据集中 age 列的第 290 到 310 行的数据。这样可以方便地检查插值效果。
```

```python
titanic["embark_town"] = titanic.embark_town.fillna("Southampton")
# 这行代码使用 pandas 的 fillna 函数，将 titanic 数据集中 embark_town（登船城市）列中的缺失值填充为 "Southampton"。
# 这意味着如果 embark_town 列中有任何缺失值，它们将被替换为 "Southampton"。

titanic.info()
```

```python
dropped_titanic = titanic.dropna(axis=0)
# 这行代码使用 pandas 的 dropna 函数删除 titanic 数据集中包含缺失值的行。
# axis=0 表示按行删除。如果某一行中有任何一个缺失值，该行将被删除。
# 删除后的数据存储在 dropped_titanic 数据框中。

dropped_titanic.info()
```

```python
grp_restaurant = restaurant.groupby("RestaurantID")

cat_list = []

for grp, grp_info in grp_restaurant:
    cat_dict = {}
    
    cat_dict['Restaurant'] = grp

    grp_res = grp_restaurant.get_group(grp)

    for index, row in grp_res.iterrows():
        # print(row)
        # break

        if row['MenuCategory'] not in cat_dict.keys():
            cat_dict[row["MenuCategory"]] = [row["MenuItem"]]
        else:
            if row['MenuItem'] not in cat_dict.keys():
                cat_dict[row["MenuCategory"]].append(row["MenuItem"])

        cat_list.append(cat_dict)

# cat_list
cat_df = pd.DataFrame(cat_list)
cat_df
```

```python
for row in cat_list:
    for key, value in row.items():
        item = {}
        if isinstance(value, list):
            for v in value:
                if v not in item.keys():
                    item[v] = 1
                else:
                    item[v] += 1
            row[key] = item

cat_list
```

```python
grp_restaurant = restaurant.groupby("RestaurantID")

cat_list = []

for grp, grp_info in grp_restaurant:
    cat_dict = {}
    
    cat_dict['Restaurant'] = grp

    grp_res = grp_restaurant.get_group(grp)

    for index, row in grp_res.iterrows():
        # print(row)
        # break

        if row['MenuCategory'] not in cat_dict.keys():
            cat_dict[row["MenuCategory"]] = [row["Price"]]
        else:
            if row['MenuItem'] not in cat_dict.keys():
                cat_dict[row["MenuCategory"]].append(row["Price"])

        cat_list.append(cat_dict)

# cat_list
cat_df = pd.DataFrame(cat_list)
cat_df
```

```python
for row in cat_list:
    for key, value in row.items():
        item = {}
        if isinstance(value, list):
            item['Min'] = min(value)
            item['Mean'] = min(value) / len(value), 2
            item['Max'] = min(value)
            row[key] = item

cat_list
```

```python
ing_list = []

for grp, gro_info in grp_restaurant:
    ing_dict = {}
    ing_dict["Restaurant"] = grp

    res_grp = grp_restaurant.get_group(grp)

    for index, row in res_grp.iterrows():
        # print(row['Ingredients'])
        # print(type(row['Ingredients']))
        temp = row["Ingredients"].replace("[", "").replace("]", "").replace("'", "")
        temptemp = temp.split(", ") # to convert to list

        for t in temptemp:
            if t not in ing_dict.keys():
                ing_dict[t] = 1

            else:
                ing_dict[t] += 1

    del ing_dict['confidential']
    ing_list.append(ing_dict)

ing_df = pd.DataFrame(ing_list)
ing_df
```

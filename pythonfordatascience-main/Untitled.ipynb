```python
# Q3

import csv

# 中文注释：读取 CSV 文件并将数据按列聚合到字典中
# English Comment: Read a CSV file and aggregate the data into a dictionary by columns
data_dict = {}

# 假设 CSV 文件中有两列：'Category' 和 'Value'
# Assume the CSV file has two columns: 'Category' and 'Value'
with open('data.csv', mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        category = row['Category']
        value = float(row['Value'])
        
        # 中文注释：根据类别对值进行累加
        # English Comment: Accumulate the value based on the category
        if category in data_dict:
            data_dict[category] += value
        else:
            data_dict[category] = value

# 中文注释：打印每个类别的总值
# English Comment: Print the total value for each category
for category, total_value in data_dict.items():
    print(f"Category: {category}, Total Value: {total_value}")

# ------------------------------------------------------------------------------- or

import csv

# 读取CSV文件，并将数据存入字典
# Read the CSV file and store data into a dictionary
data = {}

with open('data.csv', mode='r') as file:  # 打开CSV文件 Open the CSV file
    csv_reader = csv.DictReader(file)  # 使用DictReader读取数据 Use DictReader to read the data
    
    for row in csv_reader:  # 遍历每一行 Iterate through each row
        category = row['Category']  # 获取分类字段 Get the category field
        value = int(row['Value'])  # 获取值字段并转换为整数 Get the value field and convert it to integer
        
        if category in data:
            data[category] += value  # 如果分类已在字典中，累加值 If category exists in the dictionary, add the value
        else:
            data[category] = value  # 如果分类不在字典中，添加新项 If category doesn't exist, add a new entry

# 打印聚合后的数据
# Print the aggregated data
for category, total_value in data.items():
    print(f"Category: {category}, Total Value: {total_value}")

# ------------------------------------------------------------------------------- or

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: 读取CSV文件并加载到DataFrame中
# Step 1: Load the CSV file into a DataFrame
file_path = 'spotify.csv'  # 设置文件路径 Set file path
df = pd.read_csv(file_path)  # 使用pandas读取CSV文件 Use pandas to read the CSV file

# Step 2: 将'发行日期'转换为日期格式，处理错误并删除NaT行
# Step 2: Convert 'track_album_release_date' to datetime format, handle errors, and drop NaT rows
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], errors='coerce', format='%Y-%m-%d')
df = df.dropna(subset=['track_album_release_date'])  # 删除无效日期 Drop rows with invalid dates

# Step 3: 筛选出2017年的数据
# Step 3: Filter the data to include only records from 2017
df_2017 = df[df['track_album_release_date'].dt.year == 2017]

# Step 4: 计算相关矩阵以分析特征之间的关系
# Step 4: Calculate the correlation matrix to analyze relationships between features
correlation_matrix = df_2017[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 
                              'instrumentalness', 'liveness', 'valence', 'tempo']].corr()

# Step 5: 可视化相关矩阵为热力图
# Step 5: Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))  # 设置图表尺寸 Set figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')  # 显示相关系数的热力图 Display heatmap with correlation coefficients
plt.title('Correlation Matrix (2017)')  # 设置图表标题 Set title of the plot
plt.show()  # 展示图表 Show the plot

# Step 6: 按音乐类型分组，计算每个类型的平均值
# Step 6: Group the data by genre and calculate the average of each feature
genre_averages_2017 = df_2017.groupby('playlist_genre').agg({
    'loudness': 'mean',  # 计算响度的平均值 Calculate the average loudness
    'mode': 'mean',  # 计算模式的平均值 Calculate the average mode
    'speechiness': 'mean',  # 计算语音的平均值 Calculate the average speechiness
    'acousticness': 'mean',  # 计算声学性的平均值 Calculate the average acousticness
    'instrumentalness': 'mean',  # 计算器乐性的平均值 Calculate the average instrumentalness
    'liveness': 'mean',  # 计算现场感的平均值 Calculate the average liveness
    'valence': 'mean',  # 计算情感的平均值 Calculate the average valence
    'tempo': 'mean',  # 计算节奏的平均值 Calculate the average tempo
    'duration_ms': lambda x: x.mean() / 60000  # 将持续时间从毫秒转换为分钟 Convert duration from milliseconds to minutes
}).reset_index()

# Step 7: 将结果保存到CSV文件中
# Step 7: Save the result to a CSV file
output_file_path = 'genre_averages_2017.csv'  # 输出文件路径 Output file path
genre_averages_2017.to_csv(output_file_path, index=False)  # 保存为CSV文件 Save as CSV file

# Step 8: 打印保存文件的路径
# Step 8: Print the file path where the results are saved
print(f'Results saved to {output_file_path}')  # 打印结果文件路径 Print result file path


```

```python
# Q4
import pandas as pd
import matplotlib.pyplot as plt

# 中文注释：读取 CSV 文件并将数据聚合
# English Comment: Read the CSV file and aggregate data
df = pd.read_csv('data.csv')

# 中文注释：假设有三列：'Category'、'Subcategory' 和 'Value'
# English Comment: Assume the CSV has three columns: 'Category', 'Subcategory', and 'Value'
agg_data = df.groupby(['Category', 'Subcategory']).sum().unstack()

# 中文注释：创建多图网格
# English Comment: Create multiple plot grids
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# 中文注释：绘制每个类别的子类别值
# English Comment: Plot the subcategory values for each category
for i, category in enumerate(agg_data.columns.levels[0]):
    ax = axes[i//2, i%2]
    agg_data[category].plot(kind='bar', ax=ax)
    ax.set_title(f"Category: {category}")
    ax.set_ylabel('Total Value')

# 中文注释：显示图形
# English Comment: Show the plots
plt.tight_layout()
plt.show()

#------------------------------------------------------------------------------- or

import pandas as pd
import matplotlib.pyplot as plt

# 从CSV文件读取数据，并进行聚合与分析
# Read data from a CSV file and perform aggregation and analysis
df = pd.read_csv('data.csv')

# 按照某一列进行数据的聚合操作（比如'Category'）
# Perform data aggregation by a specific column (e.g., 'Category')
grouped_data = df.groupby('Category').sum()

# 设置绘图区域为2行2列的网格
# Set up a 2x2 grid for multiple plots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# 在每个网格中绘制不同的数据
# Plot different data in each grid
grouped_data['Value1'].plot(kind='bar', ax=axs[0, 0], title='Value1 Distribution')
grouped_data['Value2'].plot(kind='line', ax=axs[0, 1], title='Value2 Trend')
grouped_data['Value3'].plot(kind='pie', ax=axs[1, 0], title='Value3 Proportion')
grouped_data['Value4'].plot(kind='scatter', x='Category', y='Value4', ax=axs[1, 1], title='Value4 Scatter')

# 调整布局并显示图表
# Adjust layout and show the plots
plt.tight_layout()
plt.show()

#------------------------------------------------------------------------------- or

import pandas as pd
import matplotlib.pyplot as plt

# 中文注释：加载数据集
# English Comment: Load the dataset
df = pd.read_csv('coffee.csv')

# 显示原始数据
# Display the dataset
print(df)

# 中文注释：根据'product_category'和'product_type'进行分组，计算每种产品类型的总订单
# English Comment: Group by 'product_category' and 'product_type', then calculate the total orders for each product type
total_orders = df.groupby(['product_category', 'product_type']).agg(total_order=('transaction_qty', 'sum')).reset_index()

# 中文注释：显示聚合后的数据
# English Comment: Display the resulting DataFrame
print(total_orders)

# 中文注释：保存结果到 CSV 文件
# English Comment: Save the results to a CSV file
output_file_path = 'total_orders_by_product.csv'
total_orders.to_csv(output_file_path, index=False)

# 中文注释：创建多个子图，按照不同的产品类别分别显示
# English Comment: Create multiple subplots to show total orders by product category
categories = total_orders['product_category'].unique()  # 获取所有产品类别 / Get all unique product categories
num_categories = len(categories)

# 中文注释：设置图形大小和子图布局
# English Comment: Set figure size and layout for subplots
fig, axes = plt.subplots(nrows=num_categories, ncols=1, figsize=(10, num_categories*4))

# 中文注释：遍历每个产品类别并绘制相应的图表
# English Comment: Loop through each category and plot the corresponding data
for i, category in enumerate(categories):
    ax = axes[i]  # 选择当前的子图 / Select the current subplot
    category_data = total_orders[total_orders['product_category'] == category]  # 获取当前类别的数据 / Get the data for the current category
    
    # 中文注释：绘制每个产品类型的订单数量条形图
    # English Comment: Plot a bar chart for total orders by product type
    ax.barh(category_data['product_type'], category_data['total_order'], color='skyblue')
    
    # 设置标题和标签 / Set title and labels
    ax.set_xlabel('Total Orders')
    ax.set_ylabel('Product Type')
    ax.set_title(f'Total Orders for {category}')
    
# 中文注释：调整图形布局以避免重叠
# English Comment: Adjust layout to prevent overlapping
plt.tight_layout()

# 中文注释：显示所有图形
# English Comment: Show all the plots
plt.show()

# -------------------------------------------------------------------------------mutliple plot code
import pandas as pd
import matplotlib.pyplot as plt

# 中文注释：加载数据集
# English Comment: Load the dataset
df = pd.read_csv('coffee.csv')

# 显示原始数据
# Display the dataset
print(df)

# 中文注释：根据'product_category'和'product_type'进行分组，计算每种产品类型的总订单
# English Comment: Group by 'product_category' and 'product_type', then calculate the total orders for each product type
total_orders = df.groupby(['product_category', 'product_type']).agg(total_order=('transaction_qty', 'sum')).reset_index()

# 中文注释：显示聚合后的数据
# English Comment: Display the resulting DataFrame
print(total_orders)

# 中文注释：保存结果到 CSV 文件
# English Comment: Save the results to a CSV file
output_file_path = 'total_orders_by_product.csv'
total_orders.to_csv(output_file_path, index=False)

# 中文注释：获取产品类别并将其分成两部分
# English Comment: Get product categories and split them into two parts
categories = total_orders['product_category'].unique()  # 获取所有产品类别 / Get all unique product categories
split_point = len(categories) // 2  # 中间的分割点 / Split point

# 第一部分产品类别 / First half of the product categories
categories_part1 = categories[:split_point]

# 第二部分产品类别 / Second half of the product categories
categories_part2 = categories[split_point:]

# 中文注释：绘制第一个图形
# English Comment: Plot the first figure
plt.figure(figsize=(10, len(categories_part1)*4))
for category in categories_part1:
    category_data = total_orders[total_orders['product_category'] == category]  # 获取当前类别的数据 / Get the data for the current category
    plt.barh(category_data['product_type'], category_data['total_order'], label=f'Total Orders for {category}', color='skyblue')

plt.xlabel('Total Orders')
plt.ylabel('Product Type')
plt.title('Total Orders for First Half of Product Categories')
plt.legend()
plt.tight_layout()
plt.show()

# 中文注释：绘制第二个图形
# English Comment: Plot the second figure
plt.figure(figsize=(10, len(categories_part2)*4))
for category in categories_part2:
    category_data = total_orders[total_orders['product_category'] == category]  # 获取当前类别的数据 / Get the data for the current category
    plt.barh(category_data['product_type'], category_data['total_order'], label=f'Total Orders for {category}', color='lightcoral')

plt.xlabel('Total Orders')
plt.ylabel('Product Type')
plt.title('Total Orders for Second Half of Product Categories')
plt.legend()
plt.tight_layout()
plt.show()
```
